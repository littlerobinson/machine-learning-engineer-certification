> **6. Pour obtenir un deuxi√®me rendez-vous, vaut-il mieux √™tre le premier speed date de la soir√©e ou le dernier ?**

```python
contingence = pd.crosstab(df['order'], df['match'])
contingence.columns = ['match_fail', 'match_ok']
contingence
```

```python
df_contingence_ok = contingence['match_ok']
fig = px.scatter(df_contingence_ok, trendline = "ols")
fig.show()
```

Ce graphique montre bien qu'il y a corr√©lation entre l'ordre de passage et le nombre de match.
Plus nous avancons dans les dates moins nous matchons.

Probablement du fait que les candidats ont plus de r√©f√©rences et que les candidats savent de plus en plus ce qu'il recherche.

Nous allons tester l'utilisation de chi2_contingence pour voir si deux groupes (match et order) ont la m√™me distribution.

- **Hypoth√®se :**

L'ordre de passage n'a pas d'influence sur le match.

```python
from scipy.stats import chi2_contingency
res = chi2_contingency(contingence)
print(res.pvalue)
```

***Comment calculer et interpr√©ter la valeur de p ?***

La valeur de p est souvent utilis√©e dans les tests d'hypoth√®ses, tests qui vous permettent de rejeter, ou non, une hypoth√®se nulle. Elle repr√©sente la probabilit√© de faire une erreur de type 1, ou de rejeter l'hypoth√®se nulle si elle est vraie.

Plus la valeur de p est petite, plus la probabilit√© de faire une erreur en rejetant l'hypoth√®se nulle est faible. Une valeur limite de 0,05 est souvent utilis√©e. Autrement dit, vous pouvez rejeter l'hypoth√®se nulle si la valeur de p est inf√©rieure √† 0,05.

****Conclusion****

La valeur est inf√©rieur √† 5%, nous pouvons rejeter l'hypoth√®se.
